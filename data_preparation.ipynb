## train data set

```python
train_essays = pd.read_csv('/home/peng_sun2/s3shared/kaggle/llm-2023/data/train_essays.csv')
train_prompts = pd.read_csv('/home/peng_sun2/s3shared/kaggle/llm-2023/data/train_prompts.csv')
```

```python
train_ds = pd.merge(
    left = train_prompts[['prompt_id', 'prompt_name','instructions', 'source_text']],
    right = train_essays[['prompt_id', 'text', 'generated']],
    on = 'prompt_id',
)
train_ds
```

```python
train_ds = train_ds[['prompt_name','instructions', 'source_text', 'text', 'generated']]\
    .rename({
        'instructions': 'prompt',
        'source_text': 'source'}, axis = 1)
```

## Complete the source texts for PERSUADE data

```python
PERSUADE_source = pd.read_csv('/home/peng_sun2/s3shared/kaggle/llm-2023/data/persuade/sources.csv')
```

```python
PERSUADE_source
```

### Append source texts for The Face on Mars and A Cowboy Who Rode the Waves

```python
prompt_name_list = [
    'The Face on Mars',
    '"A Cowboy Who Rode the Waves"'
]
for prompt_name in prompt_name_list:
    gpt4_summary_source = PERSUADE_source.loc[
        PERSUADE_source.prompt_name == prompt_name, 'gpt4_summary'
    ].values[0]
    
    PERSUADE_corpus.loc[
        PERSUADE_corpus.prompt_name == prompt_name, 'source'
    ] = gpt4_summary_source
    
```

```python
gpt4_summary_source
```

```python
PERSUADE_corpus.prompt_name.unique()
```

### Append source texts for car-free cities and Does the electoral college work? to PERSUADE data sets

```python
prompt_name_list = [
    'Car-free cities',
    'Does the electoral college work?'
]

for prompt_name in prompt_name_list:
    print(prompt_name)
    source_text = train_ds.loc[
        train_ds.prompt_name == prompt_name, 'source'
    ].iloc[0]
    PERSUADE_corpus.loc[
        PERSUADE_corpus.prompt_name == prompt_name, 'source'
    ] = source_text
```

## Merge

```python
display(train_ds.shape)
display(PERSUADE_corpus.shape)
```

```python
train_ds = pd.concat([
    train_ds, PERSUADE_corpus
], axis = 0).reset_index(drop = True)

train_ds
```
